### Outliers Detection 

```python
data.head()
```

```python
plt.subplots(figsize=(40,40))
sns.boxplot(data)
```

```python
q1= data.MonthlyIncome.quantile(0.25)# finding q1 formula
q3= data.MonthlyIncome.quantile(0.75)# finding q3 formula
IQR = q3-q1 
IQR
upper_limit = q3+1.5*(IQR)
upper_limit
data=data[data.MonthlyIncome<upper_limit]
```

### Encoding

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data["Attrition"]=le.fit_transform(data["Attrition"])
data["Attrition"]
```

```python
data["Department"]=le.fit_transform(data["Department"])
data["Department"]
```

```python
data["EducationField"]=le.fit_transform(data["EducationField"])
data["EducationField"]
```

```python
data["BusinessTravel"]=le.fit_transform(data["BusinessTravel"])
data["BusinessTravel"]
```

```python
data["Gender"]=le.fit_transform(data["Gender"])
data["Gender"]
```

```python
data["JobRole"]=le.fit_transform(data["JobRole"])
data["JobRole"]
```

```python
data["MaritalStatus"]=le.fit_transform(data["MaritalStatus"])
data["MaritalStatus"]
```

```python
data["Over18"]=le.fit_transform(data["Over18"])
data["Over18"]
```

```python
data["OverTime"]=le.fit_transform(data["OverTime"])
data["OverTime"]
```

```python
data.head()
```

### Splitting data into dependent and independent variables.

```python
x = data.drop('Attrition', axis = 1)
```

```python
x.head(2)
```

```python
y = data['Attrition']
```

```python
y.head(2)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ms = StandardScaler()
x_scaled = ms.fit_transform(x)
```

```python
x
```

### Splitting into train and test data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
data
```

### Performance Metrics

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
pd.crosstab(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
pred=dt_model.predict(x_test)
```

```python
pred
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
pd.crosstab(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn import tree
plt.figure(figsize=(25,15))
tree.plot_tree(dt_model,filled=True)
```

```python
from sklearn.model_selection import GridSearchCV
parameter={
 'criterion':['gini','entropy'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
grid_search=GridSearchCV(estimator=dt_model,param_grid=parameter,cv=5,scoring="accuracy")
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
dtc_cv=DecisionTreeClassifier(criterion= 'entropy',
 max_depth=3,
 max_features='sqrt',
 splitter='best')
dtc_cv.fit(x_train,y_train)
```

```python
pred=dtc_cv.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
forest_params = [{'max_depth': list(range(10, 15)), 'max_features': list(range(0,14))}]
```

```python
rfc_cv= GridSearchCV(rfc,param_grid=forest_params,cv=10,scoring="accuracy")
```

```python
rfc_cv.fit(x_train,y_train)
```

```python
pred=rfc_cv.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```
